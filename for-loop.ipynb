```python
# 10. Develop a program that prints the first `n` Fibonacci numbers using a `for` loop.
def fibonacci(n):
    if n < 0:
        return []
    elif n == 0:
        return 0
    elif n == 1:
        return 0,1
    elif n > 1:
        a, b = 0, 1
        print(a, b, end=' ')
        for i in range(2, n):
            c = a + b
            print(c, end=' ')
            a = b
            b = c
            
n = 5
print(f"the first n Fibonacci numbers: \n")
fibonacci(n)
```

```python
# 12. Create a program that prints the prime numbers in a given range using a `for` loop.
def prime(r):
    li = []
    for i in range(2, r+1):
        for j in range(2, i):
            if i % j == 0:
                break
        else :
            li.append(i)
    return li

r = 20
print(f"the prime numbers in a given range: {prime(r)}")
```

```python
# 14. Write a program to find the maximum element in a 2D list using a nested `for` loop.
def find_2d(list):
    max = list[0][0]
    for row in list:
        for value in row:
            if max < value:
                max = value
    return max

list = [[2,4,2],[5,7,1],[8,6,7],[0,1,3]]
print(f"the maximum element in a 2D list: {find_2d(list)}")
```

```python
# 17. Write a program that converts a list of Fahrenheit temperatures to Celsius using a `for` loop.
def f2c(fahrenheit):
    c_list = []
    for f in fahrenheit:
        c = (f - 32) * (5/9)
        
        c_list.append(round(c,1))
    return c_list

fahrenheit = [132, 235, 342, 190, 295]
print(f"a list of Fahrenheit temperatures converts to Celsius: {f2c(fahrenheit)}")
```

```python
# 20. Write a program to find the greatest common divisor (GCD) of two numbers using a `for` loop.
def factor(num):
    f_list = []
    for i in range(1,num+1):
        if num % i == 0 :
            f_list.append(i)
    return f_list

def gcd(num1, num2):
    gcd_list = []
    for i in factor(num1):
        if i in factor(num2):
            gcd_list.append(i)
    return sorted(gcd_list)[-1]

num1, num2 = 24, 36
print(f"the greatest common divisor (GCD) of {num1} and {num2} is: {gcd(num1, num2)}")
```

```python
# 22. Write a program to find the prime factors of a given number using a `for` loop and list comprehension.
def prime_factors(num):
    factor = [i for i in range(1, num+1) if num % i == 0]
    prime = [i for i in factor if all(i % j != 0 for j in range(2, i))]
    return prime

number = int(input("enter a number: "))
print(f"the prime factors of a {number}: {prime_factors(number)}")
```

```python
# 31. Write a program to generate a list of prime numbers using the Sieve of Eratosthenes algorithm and list comprehension.
def Sieve_of_Eratosthenes(num):
    prime_list = [True for i in range(num+1)]
    prime_list[0] = prime_list[1] = False
    
    p = 2
    while(p*p <= num):
        if prime_list[p] == True:
            for i in range(p*p, num+1, p):
                prime_list[i] = False
        p += 1
        
    primes = [pri for pri in range(num+1) if prime_list[pri] == True]
    return primes

num = 100
print(Sieve_of_Eratosthenes(num))
```

```python
# 34. Write a program that calculates the mean, median, and mode of a list of numbers using list comprehension.
from statistics import mean, median, mode

def stat(a_list):
    a_mean = round((sum(a_list) / len(a_list)), 2)
    
    a_mode = max(set(a_list), key = a_list.count)
    
    middle = len(sorted(a_list)) // 2
    a_median = (sorted(a_list)[middle] + sorted(a_list)[~middle]) / 2
    
    return a_mean, a_median, a_mode

a_list = [3,1,8,3,3,6,8,2,4,5,9]
print(f"the mean, median, and mode of a list of numbers is: {stat(a_list)}")
```

```python
# 38. Create a program that filters a list of strings to include only those with more than three vowels using list comprehension.
def count_vowel(word):
    return sum(1 for char in word if char in "AaEeIiOoUu")

def make_a_list(strings_list):
    return [i for i in strings_list if count_vowel(i) > 3]

list_of_strings = ['HELLOOOO', 'souraav', "how", "are", 'youuu']
print(f"a list of strings with more than three vowels: {make_a_list(list_of_strings)}")
```
